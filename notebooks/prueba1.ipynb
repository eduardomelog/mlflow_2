```python
import mlflow
import numpy as np
import os
```

```python
# Forzamos MLflow a usar un path local y relativo al repo
mlflow.set_tracking_uri("file:../mlruns")

# Creamos / usamos el experimento
mlflow.set_experiment("experimento_2")
```

```python
for i in range(3):
    with mlflow.start_run(run_name=f"run_{i+1}"):

        lr = np.random.uniform(0.0001, 0.1)
        batch_size = np.random.choice([16, 32, 64])
        accuracy = np.random.uniform(0.7, 0.95)

        mlflow.log_param("learning_rate", lr)
        mlflow.log_param("batch_size", batch_size)
        mlflow.log_metric("accuracy", accuracy)

        print(f"Run {i+1} registrado")
```
